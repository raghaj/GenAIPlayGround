```python
import vertexai
from vertexai.language_models import CodeChatModel

vertexai.init(project="raghagai", location="us-central1")
chat_model = CodeChatModel.from_pretrained("codechat-bison@001")
parameters = {
    "max_output_tokens": 1024,
    "temperature": 0.2
}
chat = chat_model.start_chat()
response = chat.send_message("""Write a Python function to check if a number is a
perfect number""", **parameters)
print(f"Response from Model: {response.text}")
response = chat.send_message("""Write a unit test""", **parameters)
print(f"Response from Model: {response.text}")
response = chat.send_message("""Write a Dockerfile for a python web application using Django""", **parameters)
print(f"Response from Model: {response.text}")
response = chat.send_message("""Write a Dockerfile for a Java web application using Maven and Spring Boot""", **parameters)
print(f"Response from Model: {response.text}")
```
